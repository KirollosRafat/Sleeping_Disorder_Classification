```python
import pandas as pd
from functions import add_bp_category,assign_age_group
```

```python
sleep_df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
```

- Well Let's just drop person ID column 

```python
df = sleep_df.drop("Person ID",axis=1)
```

- Check unique values of each categorical column

```python
for col in df.columns:
    if df[col].dtype == 'object':
        print(f"{col} :{df[col].dtype}:{df[col].unique()}")
```

- Not too much to clean but sleep disorder has "nan" instead of None for no disorder, we need to handle such thing

```python
import numpy as np
df["Sleep Disorder"] = np.where(df["Sleep Disorder"].isna(),"NONE",df["Sleep Disorder"])
```

```python
df.isna().sum()
print("")
```

- NO missing data 

- Grouping Ages into categorical feature would be great for lower processing 

```python
import numpy as np
df['Age_Group'] = df['Age'].apply(assign_age_group)
```

- Getting a new feature out of Blood Pressure

```python
df = add_bp_category(df,"Blood Pressure","BP_Category")
```

- Re-arrange columns

```python
df = df[[*df.columns.drop('Sleep Disorder'), 'Sleep Disorder']]
```

- As a final step for data arrangement we have to drop all the coulmns of no interest such as Age,Blood Pressure,Systolic,Diastolic

```python
df = df.drop(["Age","Blood Pressure","Systolic","Diastolic"],axis=1)
```

```python
df.value_counts("Sleep Disorder",normalize=True) *100
```

- From The previous results there is no data imbalance present, We are good to go.

```python
df.to_csv("Sleep_Disorder_Preprocessed.csv",index=0)
```
